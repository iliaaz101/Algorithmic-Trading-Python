## Matplotlib: The Figure Object

3 steps for creating a plot using a Figure Object:
1. create a figure via plt.figure() and assign it to a variable
2. create a set of axes and assign it to a variable
3. plot your data on top of your axes

```python
fig = plt.figure() # this is a blank figure canvas that doesn't show b/c there's nothing there
axes = fig.add_axes([0,0,1,1])
# the first 0,0 is refering to the lower left corner 
# the second 1,1 is refering to the width and height of the axes (this is a ratio)
```

```python
# Data for us to plot out
a = np.linspace(0,10,11)
b = a**4
x = np.arange(0,10)
y = x*2
```

```python
a
```

```python
b
```

```python
x
```

```python
# ALL OF THE FOLLOWING COMMANDS MUST BE IN THE SAME JUPYTER NOTEBOOK CELL

fig = plt.figure() # create the figure

# LARGE AXES
axes1 = fig.add_axes([0,0,1,1]) # create the axes for your figure
axes1.plot(a,b) # plot your data on the axes of your figure

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# SMALL AXES
axes2 = fig.add_axes([0.2,0.5,0.25,0.25]) # IRL, this can be used to a magnification of our big data
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')
```

```python
fig = plt.figure(dpi=300,figsize=(2,2)) #figsize param is in inches

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

# we dont see the axes in the saved image because of the bounding box of the fig
# we need to set the bounding box param of the .savefig() method to tight
fig.savefig('new_figure.png',bbox_inches='tight')
```

## Matplotlib: Subplot Functionality

- plt.subplot() returns a tuble containing the Figure canvas and then a numpy array holding the axes object
    - each index is referring to its according plot
    - by convention, we label the tuple returned by plt.subplots() as **(fig,axes)**
        - **fig** is the entire Figure canvas
        - **axes** is a numpy array holding each of the axes according to position in the overall canvas

```python
a = np.linspace(0,10,11)
b = a ** 4
```

```python
x = np.arange(0,10)
y = 2*x
```

```python
fig, axes = plt.subplots(nrows=2,ncols=2) # fig,axes is for tuple unpacking

axes[0,0].plot(x,y)
axes[1,1].plot(a,b)
```

```python
# observing the documentation for .subplots(), we can see that we can also use
# any params from the figure object in the subplots object
fig, axes = plt.subplots(nrows=2,ncols=2,figsize=(4,10),dpi=300) # fig,axes is for tuple unpacking

axes[0,0].plot(x,y)
axes[0,1].plot(x,y)
axes[1,0].plot(x,y)
axes[1,1].plot(a,b)

# wspace and hspace params for .subplots_adjust() method are ratios of axes length
# fig.subplots_adjust(wspace=1, hspace=0.5)

# we can add titles and axes labels to each subplot by using the same 2D array indexing
axes[0,0].set_xlabel('X label 0,0')
axes[1,0].set_ylabel('Y label 1,0')
axes[1,1].set_title('Power of 4')
axes[0,1].set_xlabel('X label 0,1')
axes[0,1].set_ylabel('Y label 0,1')

fig.suptitle('Figure Level Title',fontsize=16) #creating a title for the whole figure including the subplots
# the .suptitle() method stands for "super title", similar to inheritance
plt.tight_layout()

fig.savefig('new_subplots.png',bbox_inches='tight')
```

## Matplotlib Styling: Legends

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# plotting more than one function on a graph, we need a legend
# For the legend to work, we must add labels to the function during the .plot() call

ax.plot(x,x,label= 'X vs X')

ax.plot(x,x**2,label='X vs X^2')

ax.legend(loc=(1.1,0.5)) # we call the .legend() method on the axes object labeled "ax"
# the loc param in the .legend() call lets us choose where the legend will show on the figure
# we can also pass in a tuple for the bottom left corner of the legend
```

## Matplotlib Styling: Colors & Styling

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# color param: choose a color
# label param: set a string label for the function, used in legend
# lw OR linewidth param: choose a line width RATIO for the function
# ls OR linestyle param: choose a line style for the function
    # possible linestyle options '--', '-', '-.', ':'
ax.plot(x,x,color='#b633bd',label='x vs x',lw=10,linestyle='--') # we can pass a RGB hex code in here

# creating our own custom linestyle
lines = ax.plot(x,x+1,color='red',label='x vs x+1',linewidth=5,ls=':')
# custom dashs sequence: [# of points you want, # point spaces you want]
lines[0].set_dashes([1, 2, 1, 2, 10, 2])
```

```python
type(lines)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

# the marker param of the .plot() method lets us set the markers for our data points
ax.plot(x,x+1,color='red',label='x vs x+1',linewidth=2,marker='o', ls='--')

# markersize or ms param: lets us set the size of our markers
ax.plot(x,x,color='purple',label='x vs x+1',linewidth=2,marker='s', ls='--', markersize=20,
       markerfacecolor='blue', markeredgewidth=3, markeredgecolor='orange')
```
